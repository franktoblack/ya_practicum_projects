# Введение



В задачи входят:

1.  Научиться прогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента;
2.  Сформировать типичные портреты клиентов: выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;
3.  Проанализировать основные признаки, наиболее сильно влияющие на отток;
4.  Сформулировать основные выводы и разработать рекомендации по повышению качества работы с клиентами:
     -  выделить целевые группы клиентов;
     - предложить меры по снижению оттока;
     - определить другие особенности взаимодействия с клиентами.

Сетью предоставлены:
-  Данные клиента за предыдущий до проверки факта оттока месяц
-  Информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента

## Модели прогнозирования оттока клиентов
[Назад к содержанию](#section_id)

```python
# Применяем модели логистической регрессии и случайного леса
X = df.drop('churn', axis = 1)
y = df['churn']

#разделяем модель на обучающую и валидационную выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 0, stratify=y)
   
#создадим объект класса StandardScaler и применим его к обучающей выборке
scaler = StandardScaler()
scaler.fit(X_train)
X_train_st = scaler.fit_transform(X_train) #обучаем scaler и одновременно трансформируем матрицу для обучающей выборки

#применяем стандартизацию к матрице признаков для тестовой выборки
X_test_st = scaler.transform(X_test)
```

```python
models = [
    LogisticRegression(solver='liblinear'),
    RandomForestClassifier()
]

def make_prediction(m, X_train, y_train, X_test, y_test):
    model = m
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    probabilities = model.predict_proba(X_test)[:,1]
    print('{:.2f} - Доля правильных ответов(accuracy)'.format(accuracy_score(y_test, y_pred)))
    print('{:.2f} - Точность(precision)'.format(precision_score(y_test, y_pred)))
    print('{:.2f} - Полнота(recall)'.format(recall_score(y_test, y_pred)))
    print('{:.2f} - F1-мера'.format(f1_score(y_test, y_pred)))
    print('{:.2f} - Качество модели(auc_roc)'.format(roc_auc_score(y_test, probabilities)))
    print()

for i in models:
    print('Значения для модели, построенной методом',i)
    print()
    make_prediction(i, X_train, y_train, X_test, y_test)
```

**Вывод:**

Показатели метрик обеих моделей близки друг к другу, но построенная методом логистической регрессии (ЛР) модель показывает себя лучше, чем модель, построенная методом случайного леса (СЛ). Обе модели согласно метрике auc_roc качественные - значение выше **0,95**
- **0.93** у ЛР против **0.92** у СЛ - Доля правильных ответов ***accuracy*** 
- **0.86** у ЛР против **0.84** у СЛ - Точность ***precision*** 
- **0.83** у ЛР против **0.82** у СЛ - Полнота ***recall*** 
- **0.85** у ЛР против **0.83** у СЛ - **F1**-мера

Рекомендации по удержанию клиентов можно на разделить на две категории: кассающиеся цен и общей привлекательности.
    
1. Ценовая политика
 - Продажа абонементов на больший срок по более выгодной цене, нежели за месяц.
 - Акция для продления абонемента по сниженной цене
 - Бонусы постоянным клиентам в виде посещения массажа или скидка в спорт-баре.

 - Расширение ассортимента дополнительных услуг.

 - Большая площадь клуба, чтобы при большом количестве людей не толпиться.
 - Широкий ассортимент оборудования для занятий и его количество, чтобы не приходилось стоять в очереди к тренажеру.
