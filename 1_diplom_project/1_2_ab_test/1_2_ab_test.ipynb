Дополнительно с помощью z-теста проверим равномерность распределения пропорций.
Проверим влияние группы В2 на А1 и влияние группы А2 на В1, то есть мы проверяем влияние второго теста на первый.

В таком случае:
- **Н0** - Статистически значимые различия в пропорциях **отсутствуют**
- **Н1** - Статистически значимые различия в пропорциях **присутствуют**

### Проверка на пересечение с маркетинговыми исследованиями
[Назад к содержанию](#section_id)

Идеальный А/B тест не должен совпадать с другими проводимыми тестами, маркетинговыми событиями и любыми меропирятиями, которые могут оказать влияние на тестируемые метрики. Но добиться такого условия практически невозможно из таких факторов, как сезонные мероприятия, праздники и т.д.

```python
# разнесём регионы по строкам
marketing_events_unwrap = (m_events.set_index(['name', 'start_dt', 'finish_dt'])
                           .apply(lambda x: x.str.split(', ')
                           .explode())
                           .reset_index().rename(columns={'regions':'region'}))

# разнесём даты по отдельным строкам из диапазона дат
marketing_events_unwrap = (marketing_events_unwrap.set_index(['name', 'region'])
                           .apply(lambda row: pd.date_range(row['start_dt'], row['finish_dt']).values, axis=1)
                           .explode().reset_index()
                           .rename(columns={0:'dt', 'name':'promo_name'}))

display(marketing_events_unwrap.head(20))    
```

На этом этапе будет проводиться оценка конверсии обоих тестируемых групп. По условиям ТЗ последовательность воронки не определена, поэтому можно считать, что она простая, без строгой последовательности и определяется самостоятельно.

Датасет f_events содержит 4 вида действий пользователей, из которых складывается воронка продаж: 
- `login` - регистрация,
- `product_page` - просмотр товара,
- `product_cart` - просмотр корзины,
- `purchase` - покупка.

Первым шагом будет изучение действий пользователей в целом: 
- Какие события воронки включены в выборку по очищенным данным, 
- Как часто они встречаются, 
- Сколько пользователей совершали каждое из этих событий 
- Какова доля пользователей, хоть раз совершивших событие. 

Построю для этого сводную таблицу, а для определения доли добавлю к ней один столбец:

```python
# создам сводную таблицу
df_event = (df.pivot_table(index='event_name', values='user_id', aggfunc=['count', 'nunique'],)
            .reset_index(level=0)
            .droplevel(1, axis=1)
            .rename(columns={'count': 'cnt_event', 'nunique': 'nunique_users'})
            .sort_values(by='cnt_event')
            .reset_index(drop=True))
# колонка cnt_event показывает и то, как часто встречаются события в выборке, и общее количество  
# пользователей, их совершивших (не уникальные действия пользователей)

df_event
```

Наиболее часто встречающее действие пользователей в выборке это регистрация (6911), а наименее - просмотр корзины (2088), что уступает непосредственно покупке (2114). Определю долю уникальных пользователей, хоть раз совершивших каждое из представленных событий:

```python
# заново считаю количество уникальных пользователей (после удаления на предыдущих этапах)
total_users = df['user_id'].nunique()
total_users
```

```python
# добавляю к сводной таблице столбец с долей пользователей, которые хоть раз совершали событие
df_event['event_user_share%'] = df_event['nunique_users']/total_users

#функция для преобразование доли в проценты
def format_perc(val):
    return f'{val:.2%}'
```

```python
# применим функцию к сводной таблице
df_event['event_user_share%'] = df_event['event_user_share%'].apply(format_perc)
df_event
```

Все участники тестирования прошли регистрацию, но только **32%** из них совершили покупку, то есть на предыдущих этапах теряется **68%** пользователей. Однако это общая картина, далее проверим конверсию по группам.

#### Конверсия группы А
[Назад к содержанию](#section_id)

```python
# создам сводную таблицу для группы А
df_event_a = (df.query('group == "A"')
              .pivot_table(index='event_name', values='user_id', aggfunc='nunique')
              .rename(columns={'user_id': 'cnt'}))
df_event_a
```

События воронки определю в нужном порядке для удобства

```python
funnel_index = ['login', 'product_page', 'product_cart', 'purchase']

df_event_a = df_event_a.reindex(funnel_index).reset_index()
df_event_a
```

Рассчитаем конверсию каждого этапа и потерю клиентов на каждом этапе. 

Создаем цикл для расчетов и сначала определяем количество потерянных клиентов на каждом этапе:

```python
# объявим переменные и пустые списки, которые будем использовать в цикле

prev_val = None #предыдущее значение
result_lost_cnt = [] #список с количеством теряемых клиентов

for index, row in df_event_a.iterrows():
    val = row['cnt']
    if prev_val is None:
        prev_val = val
    result_lost_cnt.append(val - prev_val)
    prev_val = val
df_event_a['lost_cnt_users'] = result_lost_cnt

df_event_a
```

```python
prev_val = None #предыдущее значение
first_val = None #первое значение
result_conv_iter = [] #список для конверсии каждого этапа
result_conv_total = [] #список для общей конверсии
result_lost_conv_total = [] #список для доли теряемых клиентов


for index, row in df_event_a.iterrows():
    val = row['cnt']
    lost_val = row['lost_cnt_users'] or val
    if prev_val is None:
        prev_val = val
        first_val = val
    result_conv_iter.append(val / prev_val)
    result_conv_total.append(val / first_val)
    result_lost_conv_total.append(lost_val / first_val)
    prev_val = val
    
df_event_a['conv_iter_%'] = result_conv_iter
df_event_a['conv_total_%'] = result_conv_total
df_event_a['conv_lost_total_%'] = result_lost_conv_total

df_event_a
```

В таблице представлены
- конверсия по этапам, 
- конверсия в покупателей 
- доля потерянных покупателей всего (от первоначального объема). 

Определяем долю потерянных клиентов по этапам. Для этого из конверсии по этапам вычитаю единицу:

```python
df_event_a['conv_lost_iter_%'] = df_event_a['conv_iter_%'] - 1
df_event_a = df_event_a.loc[:, ['event_name', 'cnt', 'lost_cnt_users', 'conv_iter_%', 
                                'conv_lost_iter_%', 'conv_total_%', 'conv_lost_total_%']]

df_event_a
```

Осталось только перевести все полученные значения долей в проценты для удобной визуализации данных. Используем для этого написанную ранее функцию и вызовем её в цикле:

```python
for col in ['conv_iter_%', 'conv_lost_iter_%', 'conv_total_%', 'conv_lost_total_%']:
    df_event_a[col] = df_event_a[col].apply(format_perc) #функция для преобразование доли в проценты
    
df_event_a
```

Все данные о конверсии в группе А получены, визуализирую их:

```python
fig = go.Figure()

fig.add_trace(go.Funnel(
    name = 'Конверсия группы А',
    orientation = "h",
    y = df_event_a['event_name'],
    x = df_event_a['cnt'],
    textposition = "inside",
    textinfo = "value+percent initial+percent previous"))

fig.update_layout(title="Воронка событий и конверсия группы А", width=900, height=500)

fig.show()
```

- Выделяется явление, когда конверсия на этапе перехода от просмотра корзины в покупку больше 100%, точнее - 105%. Такое возможно при наличии опции покупки товара без предварительного просмотра корзины. 

- В группе А общая конверсия по метрикам составила:
  - в просмотр карточек - 65%,
  - в просмотр корзины - 31%,
  - в покупку - 32%.

- Доля потерь клиентов на просмотрах карточек и на просмотрах корзины  находится примерно на одном уровне - 35%, а на покупке - 2%.

- С точки зрения потери клиентов по этапам, наиболее существенно падение фиксируется на этапе просмотра корзины: с 1161 до 548 человек, что составляет около 47%.

Так как у нас простая воронка и нет абсолютной зависимости между просмотром корзины и покупко, а так же количество пользователей, совершивших покупку больше, чем количество пользователей, просмотревших корзину, необходимо сосредоточиться на повышении конверсии на этапе просмотра карточки товара. Тогда больший поток клиентов увеличит конверсию на следующих этапах.

Статистически значимые различия выявляют при сравнении двух выборок в целях оценки вероятности верности выдвинутых гипотез. 

Также при поиске статистически значимых различий для получения результата о вероятности наличия ошибки в нулевой гипотезе требуется устновить критический уровень статистической значимости: значение, при превышении которого мы получим некую вероятность достоверности нулевой гипотезы, а в случае его не достижения, мы получаем некую вероятность наличия между выборками статистически значимой разницы, а следовательно, и ошибки в нулевой гипотезе.

Критический уровень статистической значимости обычно устанавливается либо равным 1%, либо равным 5%. В рамках исследования мы работаем с небольшими выборками и всего двумя группами тестирования, поэтому для чистоты эксперимента можем провести сравнения с обоими уровнями значимости.

Кроме того, необходимо помнить, что несколько сравнений, проводимых на одних и тех же группах — это уже множественный тест. Его важная особенность в том, что с каждой новой проверкой гипотезы растёт вероятность ошибки первого рода (получения ложнопозитивного результата статистического теста). Для снижения вероятности получения подобных результатов можно применить метод Бонферрони для корректировки уровня значимости.

Выдвенем гипотезы:

- **Н0** - Между долями уникальных посетителей, побывавших на каждом этапе воронки, __различий нет__.
- **Н1** - Между долями уникальных посетителей, побывавших на каждом этапе воронки, __различия есть__.


Начнем с подготовки нужного датафрейма: определю количество уникальных пользователей в каждой группе в разрезе всех событий:

```python
df_test_group = (df.pivot_table(index = 'event_name', columns='group', values='user_id',
                                    aggfunc='nunique', margins = True, margins_name='total')
                 .reset_index())

df_test_group
```

Соотношение общего количества юзеров и зарегистрированных участников теста 100%. Поле можно удалить из таблицы, так как нет смысла проверять его.


```python
df_test_group.drop(index=[0], inplace=True)
df_test_group
```

Создам функцию для проведения z-теста для поиска статистических критериев разниц между выборками (проведу проверку гипотезы о равенстве долей в выборках) по каждому событию:

```python
def z_test_group(df, event, group_a, group_b, alpha):
    def get_val(ev, group):
        return df.query('event_name == @ev')[group].values[0]
    
    successes = np.array([get_val(event, group_a), get_val(event, group_b)])
    trials = np.array([get_val("total", group_a), get_val("total", group_b)])
    
    # Пропорция успехов в первой группе:
    p1 = successes[0]/trials[0]

    # Пропорция успехов во второй группе:
    p2 = successes[1]/trials[1]

    # Пропорция успехов в комбинированном датасете:
    p_combined = (successes[0] + successes[1]) / (trials[0] + trials[1])

    # Разница пропорций в датасетах
    difference = p1 - p2 

    # Считаем статистику в ст.отклонениях стандартного нормального распределения
    z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/trials[0] + 1/trials[1]))
    
    # Задаем стандартное нормальное распределение (среднее 0, ст.отклонение 1)
    distr = st.norm(0, 1)  

    p_value = (1 - distr.cdf(abs(z_value))) * 2

    print('Z-тест для события {}, группы {} и {}'.format(event, group_a, group_b))
    print('p-значение: ', p_value)

    if p_value < alpha:
        print('Отвергаем нулевую гипотезу: между долями есть статистически значимые различия')
    else:
        print('Не получилось отвергнуть нулевую гипотезу: между долями нет статистически значимых различий')
    
    print('\n')
        
```

Для того, чтобы вручную не вносить значения в функцию, создадим переменные с нужными нам данными:

```python
# подготовим список событий, участвующих в эксперементах
test_events = df_test_group['event_name'].tolist()[:-1]

# определим общее количество экспериметов
experiments = len(test_events)
```

Напишем функцию, которая будет запускать сразу все пары тестируемых групп:

```python
def do_cases(alpha):
    print('\033[1m Проверка группы A и B \033[0m')
    print('\n')
    for event in test_events:
        z_test_group(df_test_group, event, 'A', 'B', alpha)
```
